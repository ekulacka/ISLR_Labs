## 2.3.2 Graphics

```python
import matplotlib.pyplot as plt

%matplotlib inline               
```

```python
x=np.random.randn(100)  # return a samples from the "standard normal" distribution.
y=np.random.randn(100)
```

```python
plt.plot(x,y,'g') #'g' is the color green
plt.xlabel('this is the x-axis')
plt.ylabel('this is the y-axis')
plt.title('Plot of X vs Y')
```

The plt.subplots() object will act as a more automatic axis manager.

```python
fig, axes = plt.subplots(figsize=(6,3))
axes.plot(x, y,)
axes.set_xlabel('this is the x-axis')
axes.set_ylabel('this is the y-axis')
axes.set_title('Plot of X vs Y');
```

```python
fig.savefig("filename.png") # To save a figure to a file we can use the savefig method in the Figure class
```

```python
x=np.arange(1,11)
print(x)
```

```python
import math
x = np.linspace(-math.pi, math.pi, num = 50)
print (x)
```

```python
import matplotlib.cm as cm
import matplotlib.mlab as mlab

y = x
X, Y = np.meshgrid(x,y)

# meshgrid is used to create a rectangular grid out of two given one-dimensional arrays.
# the output of coordinates by meshgrid can also be used for plotting functions within the given coordinate range.

f = np.cos(Y)/(1 + np.square(X))
Cnt = plt.contour(X, Y, f)
plt.show()
```

```python
f_1 = (f - f.T)/2 #f.T for transpose or tranpose(f)
plt.imshow(f_1, extent=(x[0], x[-1], y[0], y[-1])) 
plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, f_1)

plt.show()
```

```python
Auto[Auto.values == '?'] # The data set also includes a number of missing observations, indicated by a question mark '?'
```

```python
Auto = Auto.replace('?', np.NaN)
```

```python
Auto[Auto.isnull().values]
```

Other way is to use the option na_values that tells Python that any time it sees a particular character or set of characters (such as a question mark),it should be treated as a missing element of the data matrix.

```python
Auto = pd.read_csv('auto-mpg.csv',na_values='?')
Auto[Auto.isnull().values]
```

```python
Auto.shape #The shape tells us that the data has 398 observations, or rows, and 9 variables, or columns.
```
